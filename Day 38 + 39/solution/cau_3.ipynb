##### Vẽ biểu đồ kiểm tra dữ liệu của cột Sodium 

````python
sns.boxplot(data=data, y='Sodium')
plt.title('Biểu đồ BoxPlot kiểm tra Outlier của Sodium')
plt.figure(figsize=(3,2))
plt.show()
````

##### Nhận xét: Dữ liệu Sodium có outlier ( upper)

````python
# Loại bỏ outlier
data['Sodium'].describe()
````

````python
sodium_mean = data['Sodium'].mean()
sodium_mean
````

````python
sodium_q3 = data['Sodium'].quantile(0.75)
sodium_q1 = data['Sodium'].quantile(0.25)
sodium_iqr = sodium_q3 - sodium_q1
sodium_iqr
````

````python
upper_limit = sodium_q3 + 1.5 * sodium_iqr
upper_limit
````

````python
# Find Outlier of Sodium and remove it
data = data[data['Sodium'] < upper_limit]
data
````

##### Nhóm dữ liệu Category và đếm theo Item, cho biết mỗi nhóm có bao nhiêu

````python
group_category_item =  data.groupby(by='Category')['Item'].count().to_frame().reset_index()
group_category_item.rename(columns = {'Item':'Item_counts'}, inplace = True)
group_category_item
````

##### Vẽ biểu đồ thể hiện dữ liệu Category

````python
plt.figure(figsize=(12,6))
sns.barplot(data=group_category_item, x='Category', y='Item_counts', palette='tab10', hue='Category')
plt.title("Biểu đồ số lượng Items có trong Category")
plt.ylabel("Item Counts")
plt.xticks(rotation=30)
plt.show()
````
